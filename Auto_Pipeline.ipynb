#1. Auto ML Pipieline 

In this work auto pipeline using Pandas Profiling and Pycaret librart is performed.

```python
import numpy as np
import pandas as pd
from pandas_profiling import ProfileReport


!pip install numpy 
!pip install pycaret
from pycaret.regression import setup, compare_models, pull, save_model
```

```python
df = pd.read_csv('raw_data.csv', index_col=None)
profile = ProfileReport(df, title="Pandas Profiling Report")
profile
```

```python
def model_experiments(df,target):
    setup(df, target = target, silent = True)
    setup_df = pull()
    print('ML Experiment Settings: ')
    setup_df
    best_model = compare_models()
    compare_df = pull()
    print('Metrics of ML models tested: ')
    compare_df
    save_model(best_model, 'best_model')

model_experiments(df,'Price')
```
